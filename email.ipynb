```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

def send_email(receiver_email, message):
    # Load secrets from environment
    sender_email = os.getenv("SENDER_EMAIL")
    password = os.getenv("APP_PASSWORD")
    
    # Email details
    subject = "WUPHF"
    body = message

    # Create message
    complete_message = MIMEMultipart()
    complete_message['From'] = sender_email
    complete_message['To'] = receiver_email
    complete_message['Subject'] = subject
    complete_message.attach(MIMEText(body, 'plain'))

    # Send email using Gmail's SMTP server
    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(sender_email, password)
        server.send_message(complete_message)
        server.quit()
        print("✅ Email sent successfully!")
    except Exception as e:
        print(f"❌ Failed to send email: {e}")

send_email("harmfulact7@gmail.com", "WUPHF is in individual tool testing phase")
```
